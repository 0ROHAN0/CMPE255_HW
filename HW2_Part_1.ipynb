```python
ap_df = apriori(basket, min_support = 0.28, use_colnames = False)
ap_df
```

```python
ap_df = apriori(basket, min_support = 0.28, use_colnames = True)
ap_df
```

```python
freq1_count = 0; freq2_count = 0;freq3_count = 0; freq4_count = 0; freq5_count = 0
ap_df_itemsets = ap_df.itemsets
for itemset in ap_df_itemsets:
  k = len(itemset)
  if k == 1:
    freq1_count +=1
  elif k == 2:
    freq2_count +=1
  elif k == 3:
    freq3_count +=1
  elif k == 4:
    freq4_count +=1
  elif k == 5:
    freq5_count +=1
print("\n Frequency count having 1 item = {} \n Frequency count having 2 item = {} \n Frequency count having 3 item = {} \n Frequency count having 4 items = {} \n Frequency count having 5 items = {} \n".format(freq1_count,freq2_count,freq3_count,freq4_count, freq5_count))
```
